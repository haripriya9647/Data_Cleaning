```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[2021,2022,2023,2024]
y=[20,30,10,45]

plt.title("Class Size",fontsize=20,
family='Arial',
color='blue')

plt.grid(linestyle='--')
plt.plot(x,y)

plt.xticks(x)
plt.show()
```

```python
x = np.array([2021, 2022, 2023, 2024])
y1 = np.array([20, 30, 10, 45])
y2 = np.array([22, 35, 15, 50])

line_style = dict(
                           marker="*",
                           markersize=30,
                           markerfacecolor="cyan",
                           markeredgecolor="red",
                           linestyle="--")

plt.plot(x, y1,**line_style)
plt.plot(x, y2, **line_style)

plt.show()
```
