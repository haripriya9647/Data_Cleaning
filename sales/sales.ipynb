```python
df['Quantity Ordered']=pd.to_numeric(df['Quantity Ordered'], errors='coerce')
df['Price Each'] = pd.to_numeric(df['Price Each'], errors='coerce')
df['Order Date'] = pd.to_datetime(df['Order Date'],errors='coerce')
```

```python
df.info()
```

```python
df.drop(df[df.isnull().all(axis=1)].index, inplace=True)
```

```python
df['Quantity Ordered'].unique()
```

```python
df.isnull().sum()
```

```python
df.fillna({'Quantity Ordered': df['Quantity Ordered'].median()}, inplace=True)
df.fillna({'Price Each': df['Price Each'].median()}, inplace=True)
df.fillna({'Order Date': df['Order Date'].ffill()}, inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Order Date']=df['Order Date'].dt.date
```
