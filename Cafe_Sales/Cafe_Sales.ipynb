```python
df['Quantity']=pd.to_numeric(df['Quantity'], errors='coerce')
df['Price Per Unit'] = pd.to_numeric(df['Price Per Unit'], errors='coerce')
df['Total Spent'] = pd.to_numeric(df['Total Spent'], errors='coerce')
df['Transaction Date'] = pd.to_datetime(
    df['Transaction Date'], errors='coerce')
```

```python
df.info()
```

```python
df.fillna({'Quantity':df['Total Spent'].div(df['Price Per Unit'])},inplace=True)
df.fillna({'Price Per Unit':df['Total Spent'].div(df['Quantity'])},inplace=True)
df.fillna({'Total Spent': df['Quantity'].mul(
    df['Price Per Unit'])}, inplace=True)


df.fillna({'Quantity':df['Quantity'].median()},inplace=True)
df.fillna({'Price Per Unit': df['Price Per Unit'].median()},inplace=True)
df.fillna({'Total Spent': df['Total Spent'].median()},inplace=True)
```

```python
df.isnull().sum()
```

```python
df.fillna({'Item':'Unknown'},inplace=True)
df.fillna({'Payment Method': 'Unknown'},inplace=True)
df.fillna({'Location': 'Unknown'},inplace=True)
df.fillna({'Transaction Date': df['Transaction Date'].ffill()}, inplace=True)

for col in ['Item', 'Payment Method', 'Location']:
    df[col] = df[col].astype('category')
    df[col] = df[col].replace(['UNKNOWN','ERROR'],'Unknown')
```
