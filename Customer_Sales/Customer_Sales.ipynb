```python
df.dropna(subset=['Customer_ID'],inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Age'].unique()
```

```python
import re

def extract_age(age):
    age_num=re.findall('[0-9]+',str(age))
    if len(age_num)>0:
        return age_num[0]
    else: 
        return age

df['Age']=df['Age'].apply(lambda x:extract_age(x))
```

```python
df['Age'].unique()
```

```python
df_age=df[df['Age']!='nan years']['Age']
```

```python
age_median =int(df_age.dropna().astype('int64').median())
```

```python
age_median
```

```python
df.replace('nan years',age_median,inplace=True)
```

```python
df.fillna({'Age': age_median},inplace=True)
```

```python
df['Age'].unique()
```

```python
df.fillna({'Purchase_Amount':df['Purchase_Amount'].median()},inplace=True)
```

```python
df.fillna({'Feedback_Score': df['Feedback_Score'].mode()[0]}, inplace=True)
```

```python
df.isnull().sum()
```

```python
for col in ['Gender','City','Country']:
    df.fillna({col:df[col].mode()[0]},inplace=True)
```

```python
df.isnull().sum()
```

```python
df.fillna({'Last_Purchase_Date': df['Last_Purchase_Date'].bfill()},inplace=True)
```

```python
df.isnull().sum()
```
